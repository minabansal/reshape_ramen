```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
#import csv
csv_file = "ramen-ratings.csv"
```

```python
#store as dataframe
ramen_df = pd.read_csv(csv_file)
```

```python
ramen_df.head()
```

```python
ramen_df2 = ramen_df[["Country", "Brand", "Variety", "Stars"]]
ramen_df2.head()
```

```python
ramen_df2.dtypes
```

```python
ramen_df2 = ramen_df2.replace({"United States": "USA", "Dubai": "United Arab Emirates", "Holland":"Netherlands", "Hong Kong": "Hong Kong S.A.R., China", "UK": "United Kingdom","Taiwan": "Taiwan Province of China", "Sarawak":"Malaysia"})
ramen_df2 =ramen_df2.loc[ramen_df2['Country']!='Fiji']
```

```python
ramen_df3 =ramen_df2.loc[ramen_df2['Stars']!='Unrated']

ramen_df3.head()
```

```python
ramen_df3['Stars'] = ramen_df3['Stars'].astype(float)
ramen_df3.dtypes
```

```python
sortedramen_df = ramen_df3.sort_values(['Stars'], ascending=False).reset_index(drop=True)
sortedramen_df.head()
```

```python
# sortedramen_df.describe
```

```python
grouped_country = pd.DataFrame(ramen_df3.groupby(["Country"])["Stars"].mean())

grouped_country.sort_values('Stars').head()
```

```python
#import csv
csv_file2 = "happiness2017.csv"
```

```python
#store as dataframe
happiness_df = pd.read_csv(csv_file2)
happiness_df.head()
```

```python
new_happiness_df = happiness_df[["Country", "Happiness.Rank", "Happiness.Score", "Economy..GDP.per.Capita.", "Health..Life.Expectancy.", "Freedom"]].copy()
new_happiness_df
```

```python
new_happiness_df = new_happiness_df.replace({"United States": "USA"})
new_happiness_df.head()
```

```python
new_happiness_df = new_happiness_df.rename(columns={"Country": "Country", "Happiness.Rank": "Happiness Rank", "Happiness.Score": "Happiness Score", "Economy..GDP.per.Capita.": "GDP per Capita", "Health..Life.Expectancy.": "Life Expectancy", "Freedom": "Freedom"})
new_happiness_df.head()
```

```python
merged_df = pd.merge( grouped_country, new_happiness_df,how='left', on='Country')
merged_df
```

```python
#create database connection
connection_string = "postgres:postgres@0.0.0.0:5432/ramen_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
merged_df.to_sql(name='ramen_happiness', con=engine, if_exists='append', index=True)
```
